As can be seen with these plots, the generated distribution of stars is consistent with the theoretical plummer sphere density.  The below histogram below is generated from the counts of stars at each radius, and since all stars have a mass of M / n, the height of the histogram represents the mass of the sphere at that radius.  Further, because all of our stars are of equal mass, this histogram should also be representative of the density of the sphere.

With $M = 2x10^{11} M_{solar}$, and with $n=10^4$, this means that each point has a mass of $2x10^7 M_{solar}$.
